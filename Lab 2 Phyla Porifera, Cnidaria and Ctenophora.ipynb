# Phylum Cnidaria
Diversity:  approx 11,000 extant species  
Date from the Precambrian  
Range in size from microscopic to jellyfish 2 meters across with tentacles 25 m long  

Important characteristics:
- **Cnidae** 
- **Mouth surrounded by tentacles**
- **Diploblastic**
- **Radially symmetrical**
- **Alternation of generations**
- **Typically have planular larvae**
- No circulatory system
- No CNS
- Nerve system is simple nerve net
- Gut (**coelenteron**) is saclike or branched, has single opening which serves as both mouth and anus.

#### <u>Order Chondrophora</u>  
*<u>Velella</u>*<u> (by the wind sailor)</u>  

Unfortunately our specimens dried up, so no drawings of these are required. These are commonly found on the Oregon Coast, so I hope to get some before the semester is over.  They are extraordinarily cool in my opinion.   

*Velella*  are pelagic hydrozoans that float on the surface and capture the wind with a sail to move them through the water.  *Velella* float by a combination of water-repellent mucus produced by glands around the base of the sail and air-filled float chambers.  The sail is composed of a material that has structural properties not unlike those of hyaline cartilage.  The stiffness of the material is somewhere between that of rubber and that of insect cuticle.  The sail's resistance to bending is due to the shape of the sail and the thickened ridges that reinforce it.  Oddly enough there are left-sailing and right-sailing morphs of *Velella*.  Rather than sailing downwind, *Velella* sail at an angle of up to 45 degrees to the wind, either left or right. One explanation for this strange distribution of morphs is that the counter-clockwise circulation of winds in this region tends to push the left sailing morphs closer to shore and the right sailing forms further away.

### Class Scyphozoa  jellyfish

Diversity: approx 200 species in 4 orders  
exclusively marine  

Important characteristics:
* **Medusoid stage predominates**
* Polyps produce medusa by asexual budding (**strobilation**)
* **Coelenteron divided by longitudinal septa**, medusa lacks a velum, typically has a **thick mesogleal layer**, distinct pigmentation
* Sense organs occur in notches alternating with tentacles
* Cnidae present in epidermis and gastrodermis
* Usually **without a ring canal**

**<u>Drawing 3 - Aurelia.</u>**  
- *Aurelia* lives in coastal waters and feeds on plankton.  It often occurs in great numbers close to shore and thus is easy to collect for invertebrate zoology classes.   

- Immerse a specimen of the moon jellyfish, *Aurelia*, in water in a blue dissecting tray or a watch glass. Most of the characteristics to see are visible without magnification, but you can also look at them under the dissecting scope. 

- Find the parts labeled in the figure below and draw and label the parts of an oral view of the specimen. Make sure the specimen is completely immersed and add a drop of methylene blue to improve the contrast.  You might try placing the drop in the oral cavity and watching where it diffuses.  

- Notice *Aurelia* lacks a velum.  In medusa that have a velum it serves to decrease the size of the opening where water is forced out, thus increasing the propulsive force.  Would you expect *Aurelia*  capable of speedy evasive maneuvers? 

- How does *Aurelia* eat?  Swimming and eating are inseparable for this creature because it feeds by filtering plankton from the water with its fringe of tentacles or trapping them in mucus secreted on its subumbrellar surface.  The oral arms then sweep the trapped plankton off the margins and into the mouth.  

![image6][6]

[6]: https://datasci.watzek.cloud/shared/bio212/images/lab2/image6.png

- Swimming is accomplished by contraction of muscle fibers that cause the bell to contract and force water out.  How does the bell contract?  *Aurelia* has eight pacemaker ganglia that are each able to generate an impulse that contracts the bell.  When all ganglia are intact, the bell contracts in a rhythmic fashion in stable cycles.  When one ganglion generates an impulse it causes the whole bell to contract and resets all the other ganglia so they can't fire for a refractory period.  If you remove all the ganglia except one, the medusa can still swim but the pattern of contractions becomes irregular.  Where do you think the ganglia are located? How does the bell expand?  The mesoglea of *Aurelia* is composed of a protein and polysaccharide matrix that is slightly less dense than sea water and serves two major swimming functions.  First, it gives the medusa a bit of buoyancy.  Secondly, its resilience causes the bell to reform its pre contraction shape after relaxation of the muscle fibers.  The expansion of the bell is much slower than the contraction, helping the medusa to swim in an aboral direction. 

- What is the function of the radial and ring canals and how do they work?  If fluid is circulated in them, what is the pattern of circulation and where does fluid enter and exit the system?

### Class Anthozoa sea anemones, corals, sea pens

Diversity:  6,000 sp in 3 subclasses (~70% of all Cnidarians)
exclusively marine  

Important characteristics:  
* Solitary or colonial
* **Without medusoid stage**
* **Cnidae epidermal and gastrodermal**
* Divided by **longitudinal mesentaries**
* Mesenchyme thick
* Tentacles - 8 or multiples of 6 and contain extensions of the coelenteron
* Have **siphonoglyph**
* Polyps reproduce sexually and asexually  

**Subclass Octocorallia** (8 mesenteries) sea fans, sea pansies, soft corals  
**Subclass Hexacorallia** (mesentaries in multiples of 6) sea anemones, true corals

* *Metridium* movement—The handout figure below shows the range of shapes *Metridium* can adopt. *Metridium* can deform its shape by contracting the longitudinal, circular and radial retractor muscles or inflating itself by pumping water into the gastrovascular cavity through the siphonoglyph.  If the *Metridium* closes its mouth and inflates itself with the siphonoglyph, in about an hour it will look like it will be rounded, and if the circular muscles are contracted when the longitudinal muscles are contracted it will be elongate and narrow.  If disturbed, *Metridium* can protect its tentacles or completely retract .  


* The properties of *Metridium* mesoglea in conjunction with the musculature allow the animal to assume these shapes.  *Metridium&  mesoglea is composed of a collagenous protein loosely imbedded in a matrix of polysaccharides and is a viscoelastic fluid.  It can be stretched up to three times its initial length but returns to its original shape when unstressed.  It resists deforming under brief forces (like waves) yet deforms under a small constant force acting over a longer period of time (like the pumping action of the siphonoglyph).  This is why something as feeble as the pumping of water into the gastrovascular cavity with flagella can serve to inflate *Metridium*.

# Phylum Ctenophora comb jellies, sea gooseberries, sea walnuts
Diversity:  ~ 200 species described in two classes. 

Important characteristics:  
* **Diploblastic (or tiploblastic)** with ectoderm and endoderm separated by a cellular mesenchyme
* **Biradial symmetry**
* **Colloblasts** (adhesive structures)
* Gut with stomodeum and canals that branch complexly throughout body
* Gut ends in two small anal pores
* No gas exchange, excretory or circulatory systems
* Nervous system in form of nerve net or nerve plexus
* Monomorphic
* **8 rows of ciliary plates** at some life stage
* Most hermaphroditic.
