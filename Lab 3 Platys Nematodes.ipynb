Today's lab introduces you to the world of animal diversity that is bilaterally symmetrical. We begin with Platyhelminthes, a phylum that includes taxa that are typically supported as the basal clade of **Bilateria**.  I include nemertines here because they look remarkably like Platyhelminthes, but they are entirely unrelated.  This lab also introduces you to the first of three major clades of Bilateria – the **Ecdysozoa**. Since Ecdysozoans include arthropods, we will dwell in this clade for a relatively long time.  

This group was unrecognized before molecular phylogenetics consistently supported the lineage as monophyletic. The group is united by having a non-cellular chitin that is molted (the molting process=”ecdysis”). The central Ecdysozoan phylum of focus today is Nematoda. A unifying theme between Platyhelminthes and Nematodes is that these groups include a remarkable amount of parasitic lineages.

# PHYLUM PLATYHELMINTHES
### *This phylum has been split into two groups, Acoelomorpha is supported as the basal lineage of bilateria, and Platyhelminthes is a phylum of lophotrochozoa.* 

Diversity:  4 classes, about 34,000 species  
free-living and parasitic flatworms inhabiting a wide range of environments  

Important characteristics:
- Unsegmented
- **Triploblastic**
- **Acoelomate**
- **Bilaterally symmetrical**
- Flattened dorsoventrally
- Complex, incomplete gut usually present (except Cestoda)
- **Simple cephalization** with central nervous system 
- **Protonephridia** present 
- **Hermaphroditic**

### CLASS TURBELLARIA - free-living flatworms  
Diversity:  approx. 3,000 species in 12 orders  
predominantly free-living and aquatic, many marine  

Important characteristics:
- Mouth leads to **stomodeal pharynx** 
- Epidermis cellular and usually **ciliated**
- **Rhabdoids** present in most

# PHYLUM NEMERTEA
Nemertine worms. These are likely “lophotrochozoa”. More on this group later.   
Diversity ~ 1150 species  

Small thin, unsegmented worms that are mostly marine and can be very long. There are some species in tropical wet environments. We will likely see some of these on the coast. Nemertine worms superficially look like Platyhelmenthes, but nemertines differ in having a circulatory system. You won’t see a blood vessel in a platyhelmenthes! But, the blood doesn’t circulate in one direction. They have no heart. They also have a one-way (flow through) digestive tract.  
See images of these in *Chapter 11.*

# PHYLUM NEMATODA
Diversity:  enormous amount - over 16,000 species described, estimated over 1 million living species.  
All have a tubular, round body form.  
Free-living or parasitic; marine, freshwater, and terrestrial.  

**Important Characteristics:**
- Bilateral, unsegmented, pseudocoelomates
- **Body round in cross section**, covered with layered cuticle
- **Amphids**-sense organs
- **Gut complete**
- Mouth surrounded by **six lips bearing sense organs**
- Only **longitudinal muscles**
- Gonochronistic

**Ascaris**
* **Habitat:**  Adults live in the small intestine of mammals.  Females may lay more than a thousand eggs a day that pass with the host's feces.  Within the egg, larva molt once and hatch when the egg is swallowed.  Once inside a new host, the larva bores through the intestinal wall and into the bloodstream.  When it is about 5 days old it bursts into the lung cavity from the lung capillaries.  After one week, during which time it molts twice, the larva crawls up the trachea and down the digestive tract to the small intestines.  Here it matures for a month before its fourth and final molt to adulthood
While Ascaris lumbricoides var. suis lives in pigs and does not infect humans, its close relative Ascaris lumbricoides var. humanis  lives in humans.  

**Diet:**  Digested material in lumen of small intestines.  
**Reproduction:**  Two sexes mate in small intestines of host.  

**Dissection**	
- Get a **pair of gloves**, dissecting equipment, a dissecting tray and pins, and a male and female Ascaris.  The male is smaller and has a hooked tail.  Locate the anterior end of the female.  It has three lips; one dorsal and two ventral.  Open the body cavity with a **longitudinal, mid-dorsal** cut and use pins to spread the integument.

# PHYLUM NEMATOMORPHA
Diversity: 325 species  
Habitat: Mostly freshwater (5 from marine)  

**Important characteristics:**
- Free-living, aquatic pseudocoelomates – nematode like.
- Gonochoristic, collagenous external cuticle that is molted
- Differ from nematodes in having no excretory system, in details of nervous & reproductive systems, and they do not have eutely.
